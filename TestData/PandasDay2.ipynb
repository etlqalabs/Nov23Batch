```python
# Group by usage 
df = pd.read_csv('emp_copy.csv')
```

```python
df
```

```python
#SQL query : select deptno,count(deptno) from emp group by deptno
df.groupby('deptno')['deptno'].count()
```

```python
#SQL query : select deptno,sum(salary) from emp group by deptno
df.groupby('deptno')['salary'].sum()
```

```python
#SQL query : select deptno,salary,count(salary) as count_salary from emp group by deptno,salary
df_group = df.groupby(['deptno','salary'])['salary'].count()
```

```python
df_group
```

```python
type(df_group)
```

```python
df_group
```

```python
df_group = df.groupby(['deptno','salary'])['eno'].count().reset_index()
```

```python
df_group.rename(columns={'eno':"salary_count"},inplace=True)
```

```python
df_group
```

```python
type(df_group)
```

```python
# use of multiple aggregate function
# SQL equivalent : select deptno,salary,sum(salary),max(salary),min(salary),count(salary) from emp group by deptno,salary
df_agg = df.groupby(['deptno','salary'])['salary'].agg(['sum','max','min','count'])
```

```python
df_agg
```

```python
df_agg.rename(columns={'sum':"total_salary","max":"Max_salary","min":"Min_salary","count":"count_salary"},inplace=True)
```
